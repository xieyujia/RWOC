```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
```

```python
ns=np.array([100,150,200,300])
meanns=np.array([6.865174,7.165786,6.887718,7.160552174])
stdns=np.array([0.973797556,0.953497101,1.228078804,1.139078733])
```

```python
fig, ax = plt.subplots(1,1, figsize=(8,8))
ax.plot(ns,meanns,'o-',label='proposed')
ax.fill_between(ns,
                meanns-stdns,
                meanns+stdns,
                alpha=0.3)
```
